```python
features=['long_hair','forehead_width_cm','forehead_height_cm','nose_wide','nose_long','lips_thin','distance_nose_to_lip_long']


x=df[features]
y=df['gender']

```

```python
train_x = x[:4000]
train_y = y[:4000]

test_x = x[4000:]
test_y = y[4000:]
```

```python
clf = clf.fit(train_x, train_y)
```
